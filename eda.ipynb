```python
# Aseguramos que los datos estén limpios
df['City'] = df['City'].astype(str).str.strip()

# Obtenemos el conteo de personas por ciudad
conteo = df['City'].value_counts()

# Nos quedamos con las ciudades que tienen al menos 2 personas
ciudades_validas = conteo[conteo >= 4].index

# Filtramos el DataFrame original
df = df[df['City'].isin(ciudades_validas)]
df['City'].value_counts()
```

```python
df.groupby('Profession').size()
```

```python
df = df[df['Profession'] == 'Student']
```

```python
df.head()
```

```python
df.groupby('Academic Pressure').size()
```

```python
reemplazos = {
    0.0: 'Ninguna',
    1.0: 'Poco',
    2.0: 'Normal',
    3.0: 'Un poco más de lo normal',
    4.0: 'Presión',
    5.0: 'Mucha presión'
}

# Reemplazamos los valores
df['Academic Pressure'] = df['Academic Pressure'].replace(reemplazos)

# Convertimos la columna a tipo objeto (texto)
df['Academic Pressure'] = df['Academic Pressure'].astype('object')

# Verificamos
df.groupby('Academic Pressure').size()
```

```python
df = df.drop(columns=['Work Pressure','Job Satisfaction'])
```

```python
df.head()
```

```python
# Aseguramos que la columna esté en formato numérico
df['CGPA'] = df['CGPA'].astype(str).str.replace(',', '.').astype(float)

# Redondeamos al entero más cercano
df['CGPA'] = df['CGPA'].round().astype('int')
df.head()
```

```python
# Aseguramos que la columna esté en formato numérico
df['Study Satisfaction'] = df['Study Satisfaction'].astype(str).str.replace(',', '.').astype(float)

# Redondeamos al entero más cercano
df['Study Satisfaction'] = df['Study Satisfaction'].round().astype('int')
df.head()
```

```python
df['Work/Study Hours'] = df['Work/Study Hours'].astype('int')
```

```python
df.groupby('Financial Stress').size()
df = df[df['Financial Stress'] !='?']
```
